```python
def buscar_descricao(navegador, url:str):
    navegador.driver.get(url)
    source = navegador.driver.page_source
    soup = BeautifulSoup(source, "lxml")
    texto = BeautifulSoup(soup.find("meta", property='og:description')["content"]).text 
    return texto
```

```python

def processar_url(url):
    # Escolhe um navegador do pool
    navegador = navegadores[hash(url) % len(navegadores)]
    try:
        descricao = buscar_descricao(navegador, url)
    # return f"URL: {url}, Descrição: {descricao}"
        return {"url":url, "descricao":descricao}   
    except Exception as err:
        print(err )
        return {"url":url, "descricao":err}   
        pass 

lista = []
# Processando as URLs usando ThreadPoolExecutor
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
    results = executor.map(processar_url, lista_urls)

    # Exibindo os resultados
    for resultado in results:
        lista.append(resultado)
        print(resultado)
```

```python
def descricao_clean(row):
    return unidecode(str(row).upper())

itens_filtrados = (self.lista_itens
 .merge(pd.DataFrame(lista), on=["url"], how="left")
 .query("~descricao.isna() and descricao != '' ")
 .assign(descricao_search = lambda df : df["descricao"].apply(descricao_clean))
 .query("descricao_search.str.contains(@regex) == False")
 )#[1:2]["descricao_search"].values
```
